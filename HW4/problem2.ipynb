```python
import cmdstanpy
from cmdstanpy import CmdStanModel
import numpy as np
```

```python
data = {
    "N_old": 20,
    "N_new": 20,
    "hits": 12,
    "misses": 8,
    "FA": 3,
    "CR":17   
}
```

```python
model_1HT = cmdstanpy.CmdStanModel(stan_file="./model/1ht.stan")

# Fit model
fit_1HT = model_1HT.sample(
    data=data,
    chains=4,
    iter_warmup=1000,
    iter_sampling=2000,
    seed=42
)
```

```python
fit_1HT.summary()
```
